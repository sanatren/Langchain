### OpenAI Embeddings

```python
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
api_key2 = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(api_key=api_key2 , model = "text-embedding-3-large")
embeddings
```

```python
text = "Let's see how this works for embeddinsg"

result = embeddings.embed_query(text)
result
```

### vector store and vector db

```python
from langchain_community.vectorstores import Chroma
# Initialize a Chroma database
db = Chroma(embedding_function=embeddings_1024)
print("Chroma database initialized.")
```

```python
# Add documents to the Chroma database
db.add_documents(documents=chunks)
print("Documents added to Chroma database.")
```

```python
# Perform a similarity search to test the database
query = "Welcome to Smallpdf"
results = db.similarity_search(query=query, k=3)
print("Search results:", results)
```
